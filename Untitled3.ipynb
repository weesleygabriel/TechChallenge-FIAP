```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

ticker = '^BVSP'
data = yf.download(ticker, start='2022-01-01', end='2024-01-01')
data.reset_index(inplace=True)

# Gerando os dias que não geraram valores reais na bolsa
index_dates = pd.date_range(start='2022-01-01', end='2024-01-01', freq='D')
df_index = pd.DataFrame(index=index_dates, columns=['dummy_column']).reset_index()
df_index.columns = ['Date', 'dummy_column']  # Renomeia a coluna para 'Date'

# Concatenando usando a coluna 'Date' como chave e mantendo os dados da nossa base original
df_concatenated = pd.merge(df_index, data, on='Date', how='left')

# Plotar série temporal original sem ser tradada - aqui analisem o comportamento com as "lacunas" não preenchidas com os dados.
plt.figure(figsize=(10, 6))
plt.plot(df_concatenated['Close'], label='Original')
plt.title('Série completa com dias sem valores')
plt.legend()
plt.show()

# Realizando o tratamento com inferência estatística

df_concatenated['Date'] = pd.to_datetime(df_concatenated['Date'])

# Agrupa os dados por semana e calcula a mediana (aqui podemos utilizar a média também...)
mediana_semanal = df_concatenated.groupby(df_concatenated['Date'].dt.week)['Close'].median()

# Preenche os valores nulos na coluna original com as medianas semanais correspondentes
df_concatenated['Close'] = df_concatenated['Close'].fillna(df_concatenated['Date'].dt.week.map(mediana_semanal))

# Plotar série temporal tratada
plt.figure(figsize=(10, 6))
plt.plot(df_concatenated['Close'], label='Série tratada com a mediana')
plt.title('Série completa com a mediana dos valores')
plt.legend()
plt.show()
```

```python
df_concatenated.index = pd.to_datetime(df_concatenated.Date, format = '%d-%m-%Y')
df_concatenated.drop("Date", inplace =True, axis=1)
df_concatenated.head()
```

```python
plt.plot(df_concatenated.index, df_concatenated.Close)
```

```python
#Testando a hipótese de normalidade dos dados
from scipy.stats import shapiro
stat, p = shapiro(df_concatenated["Close"])
print ("Estatistica de teste: {:.4f}, valor p: {}".format(stat,round(p, 5)))
if p > 0.05:
  print ('Não há evidencia o suficiente para rejeitar a hipótese de normalidade')
else:
  print('A hipótese de normalidade é rejeitada')
```

```python
#Decompondo a base
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df_concatenated['Close'])

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(10, 8), sharex=True)

result.trend.plot(ax=ax1, label='Tendência')
ax1.set_title('Decomposição Sazonal - Tendência')

result.seasonal.plot(ax=ax2, label='Sazonalidade')
ax2.set_title('Decomposição Sazonal - Sazonalidade')

result.resid.plot(ax=ax3, label='Resíduos')
ax3.set_title('Decomposição Sazonal - Resíduos')

df_concatenated['Close'].plot(ax=ax4, label='Original', color='black')
ax4.set_title('Série Temporal Original')

plt.show()
```

```python
#Definindo se a série é estacionária ou não
import seaborn as sns
from statsmodels.tsa.stattools import adfuller

sns.set_style('darkgrid')

X = df_concatenated['Close']
result2 = adfuller(X)

print(f'ADF Statistic: {result2[0]}')
print(f'P-value: {result2[1]}')

print(f'Critical Values:')
for key, value in result2[4].items():
    print('\t%s: %.3f' % (key, value))

if result2[0] < result2[4]['5%']:
    print('A série é estacionária com 95% de confiança.')
else:
    print('A série não é estacionária.')
```
